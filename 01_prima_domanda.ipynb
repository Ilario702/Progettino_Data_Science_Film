[Come è variato l'incasso del cinema e dei generi negli anni?]

prima in generale
poi metti i generi che sono variati di più e quelli di meno

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_pickle('Film.pickle')
```

```python
# Inizializzazione dello stile
plt.style.use('seaborn-v0_8-whitegrid')

# Calcolo il prodotto lordo per ogni anno
dfTest = df.groupby('year')['lordo'].sum()

plt.figure(figsize = (13, 5))
ax = sns.lineplot(x = dfTest.index , y = dfTest.values, linestyle='-', linewidth='1', marker='o')

# Formatto in modo che mostri come scala i miliardi
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x * 1e-9:.0f}'))

# Griglia
ax.grid(True, color = "#BBBBBB", linewidth = 0.5)
ax.grid(True, axis = 'x', which = 'minor', alpha = 0.3)
ax.grid(True, axis = 'y', which = 'minor', alpha = 0.3)

# Assi
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_minor_locator(plt.MultipleLocator(1))
ax.yaxis.set_major_locator(plt.MultipleLocator(1_000_000_000))
ax.yaxis.set_minor_locator(plt.MultipleLocator(500_000_000))

# Titoli degli assi
ax.set_title('Andamento degli incassi totali del cinema negli anni', fontsize=25, fontweight='bold', pad=20)
ax.set_xlabel('Anno', fontsize=20, fontweight='normal', labelpad=15)
ax.set_ylabel('Incassi totali (miliardi di dollari)', fontsize=20, fontweight='normal', labelpad=15)
ax.set_xlim(1994, 2019)

# Salvataggio
plt.savefig("figures/1_1.png", bbox_inches='tight')
plt.savefig("figures/1_1.svg", bbox_inches='tight')
```

```python
# Grafico
plt.figure(figsize=(14, 8))
ax = sns.lineplot(data=df, x='year', y='lordo', hue='genere', marker='o', palette='tab20')

# Formatto in modo che mostri come scala i miliardi
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x * 1e-9:.0f}'))

# Titoli degli assi
plt.title('Andamento degli incassi negli anni suddiviso per genere', fontsize=25, fontweight='bold', pad=20)
plt.xlabel('Anno', fontsize=20, fontweight='normal', labelpad=15)
plt.ylabel('Incassi (miliardi di dollari)', fontsize=20, fontweight='normal', labelpad=15)

# Griglia
ax.grid(True, color = "#BBBBBB", linewidth = 0.5)
ax.grid(True, axis = 'x', which = 'minor', alpha = 0.3)
ax.grid(True, axis = 'y', which = 'minor', alpha = 0.3)

# Tacche degli assi
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_minor_locator(plt.MultipleLocator(1))
ax.yaxis.set_major_locator(plt.MultipleLocator(1_000_000_000))
ax.yaxis.set_minor_locator(plt.MultipleLocator(500_000_000))

# Lehenda
plt.legend(title='Genere', frameon=True, facecolor='white', edgecolor='gray', bbox_to_anchor=(1.05, 1), loc='upper left')

# Salvataggio
plt.savefig("figures/1_2.png", bbox_inches='tight')
plt.savefig("figures/1_2.svg", bbox_inches='tight')
```
