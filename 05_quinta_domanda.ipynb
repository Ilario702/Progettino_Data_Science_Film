confrontare il numero di biglietti venduti dei film dell'anno tra categorie (box plot)
[quanti biglietti vendono i migliori film, di ogni genere?]
[Quali sono i generi che hanno i migliori film dell'anno che vendono più biglietti?]
Qual è il genere i cui migliori film dell'anno hanno venduto più biglietti?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

df = pd.read_pickle('Film.pickle')
# Inizializzazione dello stile
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
plt.figure(figsize=(16, 8))
ax = sns.boxplot(data=df, x='genere', y='biglietti_venduti', hue='genere', palette='tab20')

# Ruoto le label dell'asse x
plt.xticks(rotation=45, ha='right')

ax.grid(True, color = "#BBBBBB", linewidth = 0.6)
ax.grid(True, axis = 'y', which = 'minor', alpha = 0.3)

# Assi
ax.yaxis.set_major_locator(plt.MultipleLocator(100_000_000))
ax.yaxis.set_minor_locator(plt.MultipleLocator(50_000_000))
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x * 1e-6:.0f}'))

# Titoli degli assi
ax.set_title('Biglietti venduti dal miglior film dell\'anno per genere', fontsize=25, fontweight='bold', pad=20)
ax.set_xlabel('Genere', fontsize=30, fontweight='normal', labelpad=15)
ax.set_ylabel('Biglietti venduti (milioni)', fontsize=20, fontweight='normal', labelpad=15)

# Salvataggio
plt.savefig("figures/5_1.png", bbox_inches='tight')
plt.savefig("figures/5_1.svg", bbox_inches='tight')
```
